Let's do PCA on this dataset to remove these correlations

```python
# Let's do PCA:
# HINT: I suggest you attempt Q5 of the quiz and then use the function you developed
#       in this and some of the following cells.

# 1. Standardise your data using a StandardScaler. Make sure you're not scaling the 'y'!

# YOUR CODE HERE

# 2. Do PCA with 2 components. How much variance is retained from the original dataset?
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
def std_pca(data, var=0.5): # define inputs: dataframe and variance
    # Your code here
    new_data = data.copy()
    for col in data.columns:
        if col == "y":
            new_data = data.drop('y',axis=1)
    std= StandardScaler().fit_transform(new_data)
    pca = PCA(n_components=var, random_state=100)
    pca_d= pca.fit_transform(std)
    print("varience retained =",sum(pca.explained_variance_ratio_))

    
    for col in data.columns:
        if col == "y":
            return np.concatenate((pca_d,data['y'].values.reshape(-1,1)),axis=1)
    return pca_d

# YOUR CODE HERE
```

## Clustering

Let's see how many clusters does K-Means find.



```python
# Use the elbow method to select the number of clusters in your data using 3 PCs.



inertias = []
for k in range(1, 9):
    kmeans = KMeans(n_clusters=k)
    y_pred = kmeans.fit_predict(std_pca(x,var=3))
    inertias.append(kmeans.inertia_)
# Let's plot inertia vs number of clusters
plt.figure(figsize=(8,5))
plt.plot(range(1, 9), inertias, 'o-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.show()

# YOUR CODE HERE
```

we can choose 2 and 3 as the possible cadidates for K

```python
# Now use the silhouette score to choose between the two candidate k values from the previous cell
sil = []
for k in range(2, 9):
    kmeans = KMeans(n_clusters=k)
    y_pred = kmeans.fit_predict(std_pca(x,var=3))
    sil.append(silhouette_score(std_pca(x,var=3), y_pred))
# Let's plot SIL vs number of clusters
plt.figure(figsize=(8,5))
plt.plot(range(2, 9), sil, 'o-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.show()

# YOUR CODE HERE
```

```python
# Visualise the first two pcs (using a scatterplot) using the labels from KMeans with your chosen value of k
from sklearn.metrics.cluster import completeness_score
from sklearn.metrics.cluster import silhouette_score
kmeans_blobs = KMeans(n_clusters=3)
y_pred_blobs = kmeans_blobs.fit_predict(std_pca(x,var=3))
plt.figure(figsize=(8,5))
plt.title("Blobs - K-Means", fontsize=12)
plt.scatter(std_pca(x,var=3)[:, 0], std_pca(x,var=3)[:, 1], c=y_pred_blobs)
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')

```
